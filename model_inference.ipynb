```python
def predict(df, scaler, model):
    cols = df.columns
    X_test = scaler.transform(df)
    X_test = pd.DataFrame(X_test, columns=cols)

    y_pred = model.predict(X_test, steps=1)

    return y_pred
```

```python
path = '/content/drive/MyDrive/metall/Прогноз металлические заготовки/cloud/'
filename = 'for_prediction_nn.csv'

df = pd.read_csv(path + filename)
df.head()
```

```python
path_scaler = '/content/drive/MyDrive/metall/Прогноз металлические заготовки/'
scaler_file = 'scaler.save'

scaler = joblib.load(path_scaler + scaler_file)
```

```python
path_model = '/content/drive/MyDrive/metall/Прогноз металлические заготовки/'
model_file = 'my_best_model.hdf5'

model = load_model(path_model + model_file)
```

```python
y_pred = predict(df, scaler, model)
print(y_pred)
```
